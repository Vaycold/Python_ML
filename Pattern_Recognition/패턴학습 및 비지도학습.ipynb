```python
# 데이터 생성 - 비선형관계
# (x,y) = ( r*cos(theta), r*sin(theta) ) 임을 활용하자
np.random.seed(1)
def circle(r, n, sigma) :
    theta = np.random.uniform(0,2*np.pi, size=n)
    x_error = np.random.normal(0, sigma, size=n)
    y_error = np.random.normal(0, sigma, size=n)

    x_val = r*np.cos(theta) + x_error
    y_val = r*np.sin(theta) + y_error

    return x_val, y_val, np.repeat(r,n)

```

```python

n = 200
r_list = [1,3,5]
sigma = 0.2
param_list = [(r,n,sigma) for r in r_list]
param_list
```

```python
coordinates = [ circle(param[0], param[1], param[2]) for param in param_list]
```

```python
color_code = {1 :'Orange', 3:'Skyblue', 5:'Green'}
```

```python
plt.figure(figsize=(7,7))
for j in range(0,len(coordinates)) :
  x,y,group = coordinates[j]
  sns.scatterplot(x,y,c=[color_code.get(i) for i in group])

plt.show()
```

```python
# 데이터가 보기 힘드니 정리를 하자.

x1,x2,y_2_bf = [np.append(np.append(coordinates[0][i], coordinates[1][i]), coordinates[2][i]) for i in range (0,3)]
```

```python
x_2 = np.c_[x1,x2]
```

```python
mapping_dic = {1 : 0,
               3 : 1,
               5 : 2}
y_2_bf2 = [mapping_dic.get(i) for i in y_2_bf]

y_2 = np.array(y_2_bf2)
```

```python
# Clustering
# K- means clustering
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3, random_state=1)
km.fit(x_1)
```

```python
centers = km.cluster_centers_
y_km = km.labels_
```

```python
# plotting
plt.figure(figsize=(8,8))
plt.scatter(
    x_1[y_km == 0, 0], x_1[y_km == 0, 1],
    s = 50, c='lightgreen',
    marker = 's', edgecolor='black',
    label = 'cluster1'
)

plt.scatter(
    x_1[y_km == 1, 0], x_1[y_km == 1, 1],
    s = 50, c='orange',
    marker = 'o', edgecolor='black',
    label = 'cluster2'
)

plt.scatter(
    x_1[y_km == 2, 0], x_1[y_km == 2, 1],
    s = 50, c='lightblue',
    marker = 'v', edgecolor='black',
    label = 'cluster3'
)

plt.scatter(
    km.cluster_centers_[:,0],km.cluster_centers_[:,1],
    s = 250, c='red',
    marker = '*', edgecolor='black',
    label = 'centroids'
)

plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

```python
# 최적의 K 찾기

distortions  = []
for k in range(1,11) : # k는 1부터 10까지중에 하자
    km = KMeans(n_clusters=k,  random_state=1)
    km.fit(x_1)
    distortions.append(km.inertia_)
```

```python
distortions # 이 distortion 값은 많아질 수록 줄어드는 것이다
# 그렇다면 어디에서 급격하게 많아지는지 elbow point로 확인하면된다.
# 일단 작을 수록좋은값이긴 함.

```

```python
plt.plot(list(range(1,11)), distortions, marker='o')
plt.show()
# 기울기가 완만해진 곳에서 찾는거임.
# 이 경우에는 2가 적당함
```

```python
# 모델에 fitting시키자
color_code = {0 : 'Orange', 1:'Skyblue'}
plt.figure(figsize=(7,7))

km = KMeans(n_clusters=2, random_state=1)
km.fit(x_1)
y_km = km.labels_


sns.scatterplot(x_1[:,0], x_1[:,1], c= [color_code.get(i) for i in y_km])
plt.show()
```

```python
# PCA 를 위한 SVD 진행
# x_1 데이터의 각 변수를 평균이 0이 되도록, 만들자
mx = x_1
for i in range(x_1.shape[1]) :
  mx[:,i] = mx[:,i] - np.mean(x_1[:,i])
```

```python
# mx의 공분산행렬을 만들자
A = np.cov(mx.transpose())
```

```python
A
```

```python
# SVD를 진행해보자
aU,aD,aV = np.linalg.svd(A)
```

```python
# PC score를 구해내어서 xy에 저장 ㅣ 이건 mx 행렬에 eigenvector를 곱해서 구한 것과 같음.
xv = np.dot(mx,aV)
```

```python
plt.scatter(mx[:,0], mx[:,1], color='black')
plt.scatter(xv[:,0], xv[:,1], color='green')
origin = (0,0)

plt.quiver(origin[0], origin[1], aV[0,0], aV[1,0], color = ['r','b'], scale=3)
plt.quiver(origin[0], origin[1], aV[0,1], aV[1,1], color = ['r','b'], scale=3)

plt.show()

# 검정 : original data
# 초록 : pc score
# pc 스코어는 원래 가지고 있던 데이터의 선형적 특성을 없애줌 
```

```python
# PCA진행

from sklearn.decomposition import PCA
pca = PCA()
pca.fit(x_1)
```

```python
pc_score = pca.transform(x_1)
print(pc_score[0:5,:])
print(xv[0:5,:])
```

```python
# eigen vector
eigen_v = pca.components_.transpose()
eigen_v
```

```python
print(aV)
print(eigen_v)
```

```python
# eigen values
print(pca.explained_variance_)
print(aD)
```

```python
mx = x_1
for i in range(x_1.shape[1]) :
  mx[:,i] = mx[:,i] - np.mean(x_1[:,i])

plt.scatter(mx[:,0], mx[:,1], color='black') # centerd origin 
plt.scatter(pc_score[:,0], pc_score[:,1], color='green')
origin = (0,0)
plt.quiver(*origin, eigen_v[0,0], eigen_v[1,0], color = ['r','b'], scale=3)
plt.quiver(*origin, eigen_v[0,1], eigen_v[1,1], color = ['r','b'], scale=3)
plt.show()
```

```python
# PC를 활용한 차원 축소

label = ['PC1', 'PC2']

index = np.arange(len(label))
plt.bar(index,pca.explained_variance_)
plt.ylabel('Eigen Values')
plt.xticks(index, label)
plt.show()

# 즉 pc1만 사용해도 됨.
```
